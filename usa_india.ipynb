```python
import pandas as pd
import numpy as np
```

```python
imdb_titles = pd.read_csv('IMDb movies.csv', encoding="utf-8", na_values="Unknown", encoding_errors="replace")
imdb_ratings = pd.read_csv("IMDb ratings.csv", usecols=["imdb_title_id", "weighted_average_vote"])
imdb = imdb_titles.merge(imdb_ratings, left_on="imdb_title_id", right_on="imdb_title_id", how="inner")
imdb
```

As per the analysis made for the indian movie dataset the major columns for the analysis are:
imdb_title_id, title, year, date_published, genre, language, country, director, writer, actors, production_company,description, weighted_average_vote

```python
imdb_movies = imdb[["imdb_title_id", "title", "year", "date_published", "genre", "language", "country", "director", "writer", "actors", "production_company", "description", "weighted_average_vote"]]
imdb_movies
```

Getting data of movie which belongs to the country USA and India

```python
usa_india_movie = imdb_movies[imdb_movies["country"] == "USA"]
usa_india_movie = pd.concat([usa_india_movie, imdb_movies[imdb_movies["country"] == "India"]])
usa_india_movie
```

There is one observation which as value in year as "TV Movie 2019" changin it in to 2019

```python
usa_india_movie.loc[usa_india_movie["year"] == "TV Movie 2019","year"] = 2019
```

```python
usa_india_movie["year"] = usa_india_movie["year"].astype(int)
usa_india_movie.groupby(by="year").size()
```

As recent vieweres don't like too much to watch old movie so taking the movies from year 1990 for this dataset

```python
usa_india_movie_1990 = usa_india_movie[usa_india_movie["year"] >= 1990]
usa_india_movie_1990
```

Checking is there any duplicate ids in the dataset

```python
usa_india_movie_1990.duplicated("imdb_title_id").sum()
```

```python
usa_india_movie_1990.isna().sum()
```

Dropping values which null values in language and description

```python
usa_india_movie_1990.shape
```

```python
usa_india_movie_1990 = usa_india_movie_1990.dropna(subset=["language", "description"])
usa_india_movie_1990.isna().sum()
```

```python
usa_india_movie_1990.shape
```

total 386 movies are dropped from the dataset

And checking once again is there in duplicate ids

```python
usa_india_movie_1990.duplicated("imdb_title_id").sum()
```

```python
usa_india_movie_1990.groupby(by="language").size().sort_values(ascending=False)
```

There are 662 distinct categories in the dataset

As there are multiple languages spearated by commas so splited movies by single language and created new title id 

```python
import random

usa_indian_movies_languages = pd.DataFrame(columns=["imdb_title_id", "title", "year", "date_published", "genre", "language", "country", "director", "writer", "actors", "production_company", "description", "weighted_average_vote"])
for index, row in usa_india_movie_1990.iterrows():
    lan = str(row["language"]).split(", ")
    if len(lan) != 1:
        for i in range(len(lan)):
            new_row = row.copy(deep=True)
            while True:
                title_id = "tt" + str(random.randint(1000000, 2000000))
                if title_id not in usa_india_movie_1990["imdb_title_id"].to_list() and title_id not in \
                        usa_indian_movies_languages["imdb_title_id"].to_list():
                    new_row["imdb_title_id"] = title_id
                    break
            new_row["title"] = str(new_row["title"]) + f"({lan[i]})"
            new_row["language"] = lan[i]
            usa_indian_movies_languages = usa_indian_movies_languages._append(new_row)
    else:
        usa_indian_movies_languages = usa_indian_movies_languages._append(row)
usa_indian_movies_languages
```

After splitting by language total movies are 24288 

Checking whether there are any duplicate ids

```python
usa_indian_movies_languages.duplicated("imdb_title_id").sum()
```

Seeing the distribution using languages

```python
languages = usa_indian_movies_languages.groupby(by="language").size().reset_index(name="count")
languages = languages.sort_values(by="count",ascending=False)
languages
```

There 141 languages present in the dataset

```python
    from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming 'languages' is a pandas DataFrame with 'language' and 'count' columns
word_freq_dict = dict(zip(languages['language'], languages['count']))

# Creating a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)

# Plotting the WordCloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Movie Languages')
plt.show()
```
