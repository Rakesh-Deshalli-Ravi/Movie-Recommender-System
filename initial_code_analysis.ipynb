```python
import pandas as pd
import numpy as np
```

```python
imdb_titles = pd.read_csv('IMDb movies.csv', encoding="utf-8", na_values="Unknown", encoding_errors="replace")
imdb_titles
```

```python
imdb_ratings = pd.read_csv("IMDb ratings.csv", usecols=["imdb_title_id", "weighted_average_vote"])
imdb_ratings
```

```python
netflix_date = imdb[['date_published']].dropna()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
               'November', 'December']
netflix_date['year'] = netflix_date['date_published'].apply(
    lambda x: int(x.split("-")[0]) if x != "TV Movie 2019" else 2019)
netflix_date['month'] = netflix_date['date_published'].apply(
    lambda x: month_order[int(x.lstrip().split('-')[1]) - 1] if len(x.lstrip().split('-')) == 3 else month_order[0])
netflix_date = netflix_date[netflix_date["year"] > 2000]

df = netflix_date.groupby('year')['month'].value_counts().unstack().fillna(0)[month_order].T
plt.figure(figsize=(10, 7), dpi=200)
plt.pcolor(df, cmap='afmhot_r', edgecolors='white', linewidths=2)  # heatmap
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns, fontsize=7, fontfamily='serif')
plt.yticks(np.arange(0.5, len(df.index), 1), df.index, fontsize=7, fontfamily='serif')

plt.title('Imdb Contents Update', fontsize=12, fontfamily='calibri', fontweight='bold', position=(0.20, 1.0 + 0.02))
cbar = plt.colorbar()

cbar.ax.tick_params(labelsize=8)
cbar.ax.minorticks_on()
plt.show()
```

As there are multiple language, creating another movie dataset which contains only single language by splitting and creting with duplicate ids

```python
import random

indian_movies_languages = pd.DataFrame(columns=['imdb_title_id', 'title', 'original_title', 'year', 'date_published',
                                                'genre', 'duration', 'country', 'language', 'director', 'writer',
                                                'production_company', 'actors', 'description', 'avg_vote', 'votes',
                                                'budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore',
                                                'reviews_from_users', 'reviews_from_critics', 'weighted_average_vote'])
for index, row in indian_movies_since_2000.iterrows():
    lan = str(row["language"]).split(", ")
    if len(lan) != 1:
        for i in range(len(lan)):
            new_row = row.copy(deep=True)
            while True:
                title_id = "tt" + str(random.randint(1000000, 2000000))
                if title_id not in indian_movies_since_2000["imdb_title_id"].to_list() and title_id not in \
                        indian_movies_languages["imdb_title_id"].to_list():
                    new_row["imdb_title_id"] = title_id
                    break
            new_row["title"] = str(new_row["title"]) + f"({lan[i]})"
            new_row["language"] = lan[i]
            indian_movies_languages = indian_movies_languages._append(new_row)
    else:
        indian_movies_languages = indian_movies_languages._append(row)
indian_movies_languages
```
